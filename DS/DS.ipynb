1. La cr√©ation de raphe non-orient√© nomme G: 


Le code repr√©sente un graphe ùê∫: 
G avec des n≈ìuds ayant des coordonn√©es 2D et des ar√™tes reliant ces n≈ìuds. Les n≈ìuds sont d√©finis par leurs √©tiquettes (A, B, C, etc.) et leurs positions (x, y), tandis que les ar√™tes sont des paires de n≈ìuds indiquant les connexions. 

```python
G = {
    'nodes': {
        'A': {'x': 0, 'y': 1},
        'B': {'x': 1, 'y': 2},
        'C': {'x': 1, 'y': 0},
        'D': {'x': 2, 'y': 1},
        'E': {'x': 3, 'y': 0},
        'F': {'x': 3, 'y': 2},
        'G': {'x': 4, 'y': 1}
    },
    'edges': [('A', 'B'), ('A', 'C'), 
              ('B', 'C'), ('B', 'D'), ('B', 'F'),
              ('C', 'D'), ('C', 'E'), ('E', 'G'),
              ('D', 'E'), ('D', 'F'), ('D', 'G'), 
              ('F', 'G')
            ]
}
```

2.Cr√©ation d'une fonction <b>graphe_non_orient√©</b> qui permet d‚Äôafficher un graphe non-orient√© en utilisant matplotlib:

Cette fonction en Python utilise matplotlib pour dessiner un graphe non-orient√©. Les n≈ìuds sont repr√©sent√©s par des cercles agrandis avec leurs √©tiquettes en leur centre, et les ar√™tes sont des lignes reliant ces n≈ìuds. Les axes X et Y sont masqu√©s pour une meilleure visualisation du graphe.

```python
import matplotlib.pyplot as plt

def graphe_non_orient√©(graph):
    fig, ax = plt.subplots()
    
    # Plot nodes with larger size and numbers inside
    for node, coord in graph['nodes'].items():
        ax.plot(coord['x'], coord['y'], 'o', markersize=20, color='skyblue')  # Larger nodes and uniform color
        ax.text(coord['x'], coord['y'], f'{node}', fontsize=12, ha='center', va='center', color='black', weight='bold')
    
    # Plot edges
    for edge in graph['edges']:
        x_values = [graph['nodes'][edge[0]]['x'], graph['nodes'][edge[1]]['x']]
        y_values = [graph['nodes'][edge[0]]['y'], graph['nodes'][edge[1]]['y']]
        ax.plot(x_values, y_values, color='black')
    
    # Remove X and Y axis
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
    plt.title('Graphe non-orient√©')
    plt.show()

```

3. L'appelle de la fonction : 

```python
graphe_non_orient√©(G)
```

1. Adaptation de la fonction pour qu‚Äôelle affiche aussi un graphe orient√©.


<b>Cette fonction utilise matplotlib pour dessiner un graphe qui peut √™tre orient√© ou non, en fonction du param√®tre is_oriented. Les n≈ìuds sont repr√©sent√©s par des cercles agrandis avec leurs √©tiquettes en leur centre. Les ar√™tes sont dessin√©es comme des lignes pour les graphes non orient√©s et comme des fl√®ches pour les graphes orient√©s. Les axes X et Y sont masqu√©s pour une meilleure visualisation.<br>

```python
from matplotlib.patches import FancyArrowPatch

def graphe_orient√©(graph, is_oriented=False):
    fig, ax = plt.subplots()
    
    # Plot nodes with larger size and numbers inside
    for node, coord in graph['nodes'].items():
        ax.plot(coord['x'], coord['y'], 'o', markersize=20, color='skyblue')  # Larger nodes and uniform color
        ax.text(coord['x'], coord['y'], f'{node}', fontsize=12, ha='center', va='center', color='black', weight='bold')
    
    # Plot edges
    for edge in graph['edges']:
        x_values = [graph['nodes'][edge[0]]['x'], graph['nodes'][edge[1]]['x']]
        y_values = [graph['nodes'][edge[0]]['y'], graph['nodes'][edge[1]]['y']]
        if is_oriented:
            # Draw directed edge
            arrow = FancyArrowPatch((x_values[0], y_values[0]), (x_values[1], y_values[1]), 
                                    arrowstyle='-|>', color='black', mutation_scale=30)
            ax.add_patch(arrow)
        else:
            # Draw undirected edge
            ax.plot(x_values, y_values, color='black')
    
    # Remove X and Y axis
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
    plt.title('Graphe orient√©' if is_oriented else 'Graphe non-orient√©')
    plt.show()
```

2. l'applle de la foction <br>graphe_orient√©():</br> <br>
La fonction prend deux param√©tre le graphe et is_oriented , Ce param√®tre sp√©cifie si le graphe est orient√© (True) ou non orient√© (False).

```python
graphe_orient√©(G,True)
```
